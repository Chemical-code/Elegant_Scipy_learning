# Chapter 2 Quantile Normalization

```python
import numpy as np
from scipy import stats

def quantile_norm(X):
    """ normalization of X row for each same distribution
    
    M gene expression matrix of N samples is given
    qunatile normalization have same data range for every sample
    
    calculate average of each column to get average row
    quantile of each row is substituted with each quantile
    
    parameter
    -------------
    X: floating point 2D array (M, N)
        M: gene/properties (column)
        N: sample (row)
    
    return value
    -------------
    Xn: floating point 2D array (M, N)
        normalized data
    """
    
    # Calculation of qunatiles
    quantiles = np.mean(np.sort(X, axis=0), axis=1)
    
    # Rank calculation for each row
    # Calculate each row's rank for each measurement value
    # Smallest value becomes 1, next smallest value becomes 2, ..., Most value becomes M
    ranks = np.apply_along_axis(stats.rankdata, 0, X)
    
    # Change index to integer 1 to M-1
    rank_indices = ranks.astype(int) - 1
    
    # Indexing quantile about each rank of rank matrix
    Xn = quantiles[rank_indices]
    
    return(Xn)
```

```python
def quantile_norm_log(X):
    logX = np.log(X + 1)
    logXn = quantile_norm(logX)
    return logXn
```

## 2.2 Difference of gene expression distribution among species

```python
from scipy import stats

def plot_col_density(data):
    # Use kernel density estimation
    density_per_col = [stats.gaussian_kde(col) for col in data.T]
    x = np.linspace(np.min(data), np.max(data), 100)
    
    fig, ax = plt.subplots()
    for density in density_per_col:
        ax.plot(x, density(x))
    ax.set_xlabel('Data value (per column)')
    ax.set_ylabel('Density')
```

```python
# Before normalization
log_counts = np.log(counts + 1)
plot_col_density(log_counts)
```

```python
# After normalization
log_counts_normalized = quantile_norm_log(counts)

plot_col_density(log_counts_normalized)
```

## 2.3 Biclustering

```python
def most_variable_rows(data, *, n=1500):
    """ variable subset data of n row
    
    n = the most variable number of gene
    
    parameter
    ----------------
    data: float point 2D array
        subset data
    n: integer, selective parameter
        the number of column that will return
    
    return value
    ----------------
    variable_data: float point 2D array
        most variable n column
    """
    
    # Calculate variance through the row
    rowvar = np.var(data, axis=1)
    # Import n data that is filtered with the ascending order
    sort_indices = np.argsort(rowvar)[-n:]
    # Obtain variable data for index
    variable_data = data[sort_indices, :]
    return variable_data
```

```python
from scipy.cluster.hierarchy import linkage

def bicluster(data, linkage_method='average', distance_metric='correlation'):
    """ cluster the column and the row of the matrix
    
    parameter
    -----------------
    data: 2D ndarray
        input data for biclustering
    linkage_method: string, selective parameter
        method that will transfer 'linkage'
    distance_metric: string, selective parameter
        metric that will use for clustering
        
    return value
    -----------------
    y_rows: linkage matrix
        column clustering
    y-cols: linkage matrix
        row clustering
    """
    
    y_rows = linkage(data, method=linkage_method, metric=distance_metric)
    y_cols = linkage(data.T, method=linkage_method, metric=distance_metric)
    return y_rows, y_cols
```

## 2.4 Cluster visualization

```python
from scipy.cluster.hierarchy import dendrogram, leaves_list

def clear_spines(axes):
    for loc in ['left', 'right', 'top', 'bottom']:
        axes.spines[loc].set_visible(False)
    axes.set_yticks([])
    axes.set_xticks([])
    
def plot_bicluster(data, row_linkage, col_linkage, row_nclusters=10, col_nclusters=3):
    """ proceed biclustering, draw dendrogram and heatmap for each axes
    
    parameter
    ---------------
    data: float point array, shape(M, N)
        input data for biclustering
    row_linkage: array, shape(M-1, 4)
        linkage matrix of 'data' column
    col_linkage: array, shape(N-1, 4)
        linkage matrix of 'data' row
    n_clusters_r, n_clusters_c: integer, selective parameter
        cluster number of row and column
    """
    
    fig = plt.figure(figsize=(4.8, 4.8))
    
    # Column direction dendrogram
    # add_axes() get rectengle value to add low rank graph to figure
    # Consider both side length as 1
    # Left-downward side of figure is (0,0)
    # Parameter of add_axis() is (left, right, top, bottom)
    # To draw left(column)-dendrogram
    # Make rectangle whose left-downward side is (0.09, 0.1)
    # Width is 0.2 and height is 0.6
    ax1 = fig.add_axes([0.09, 0.1, 0.2, 0.6])
    # Obtain a part of linkage tree
    threshold_r = (row_linkage[-row_nclusters, 2] + row_linkage[-row_nclusters+1, 2]) / 2
    with plt.rc_context({'lines.linewidth':0.75}):
        dendrogram(row_linkage, orientation='left', color_threshold=threshold_r, ax=ax1)
    clear_spines(ax1)
    
    # Row direction dendrogram
    ax2 = fig.add_axes([0.3, 0.71, 0.6, 0.2])
    threshold_c = (col_linkage[-col_nclusters, 2] + col_linkage[-col_nclusters+1, 2]) / 2
    with plt.rc_context({'lines.linewidth':0.75}):
        dendrogram(col_linkage, color_threshold=threshold_c, ax=ax2)
    clear_spines(ax2)
    
    # Draw heat infrared graph
    ax = fig.add_axes([0.3, 0.1, 0.6, 0.6])
    
    # Dendrogram terminal node (leaves) data align
    idx_rows = leaves_list(row_linkage)
    data = data[idx_rows, :]
    idx_cols = leaves_list(col_linkage)
    data = data[:, idx_cols]
    
    im = ax.imshow(data, aspect='auto', origin='lower', cmap='YlGnBu_r')
    clear_spines(ax)
    
    # Axis label
    ax.set_xlabel('Samples')
    ax.set_ylabel('Genes', labelpad=125)
    
    # Graph legend
    axcolor = fig.add_axes([0.91, 0.1, 0.02, 0.6])
    plt.colorbar(im, cax=axcolor)
    
    # Display graph
    plt.show()
```

```python
counts_log = np.log(counts + 1)
counts_var = most_variable_rows(counts_log, n=1500)
yr, yc = bicluster(counts_var, linkage_method='ward', distance_metric='euclidean')
with plt.style.context('style/thinner.mplstyle'):
    plot_bicluster(counts_var, yr, yc)
```

## 2.5 Prediction of survival rate

```python
patients = pd.read_csv('data/patients.csv', index_col=0)
patients.head()
```

```python
def survival_distribution_function(lifetimes, right_censored=None):
    """ return survival distribution function of lifetimes set
    
    parameter
    ---------------
    lifetimes: float or integer array
        lifetime of observed patients (must not be negative)
    right_censored: boolean array (same shape with lifetime array)
        True = lifetime does not observed
        np.nan value of lifetimes array considered as overranged right-censored value
    
    return value
    ---------------
    sorted_lifetimes: float array
    sdf: float array
        value starts from 1 and decrease gradually
        states about each observation of lifetimes array
    """
    n_obs = len(lifetimes)
    rc = np.isnan(lifetimes)
    if right_censored is not None:
        rc |= right_censored
    observed = lifetimes[~rc]
    xs = np.concatenate(([0], np.sort(observed)))
    ys = np.linspace(1, 0, n_obs + 1)
    ys = ys[:len(xs)]
    return xs, ys
```

```python
def plot_cluster_survival_curves(cluster, sample_names, patients, censor=True):
    """ Plot the survival data from a set of sampel clusters
    
    parameter
    ----------------
    cluster: integer array(pd.Series enumerated parameter)
        cluster index of each sample
        simple integer encoded by pandas enumerated parameter
    smaple_names: string list
        name for each sample
        should be same length with cluster parameter
    patients: pandas.DataFrame
        data_frame includes survival info for each patient
        index of data_frame need to be corrected with sample_names
        sample without list will be ignored
    censor: boolean, selective parameter
        if true, use patients['melanoma-dead'] check survival data as righ-censored data
    """
    fig, ax = plt.subplots()
    if type(clusters) == np.ndarray:
        cluster_ids = np.unique(clusters)
        cluster_names = ['cluster {}'.format(i) for i in cluster_ids]
    elif type(clusters) == pd.Series:
        cluster_ids = cluster.cat.categories
        cluster_names = list(cluster_ids)
    n_clusters = len(cluster_ids)
    for c in cluster_ids:
        clust_samples = np.flatnonzero(clusters == c)
        # Remove patients who are not included to survival data
        clust_samples = [sample_names[i] for i in clust_samples if sample_names[i] in patients.index]
        patient_cluster = patients.loc[clust_samples]
        survival_times = patient_cluster['melanoma-survival-time'].values
        if censor:
            censored = ~patient_cluster['melanoma-dead'].values.astype(bool)
        else:
            censored = None
        stimes, sfracs = survival_distribution_function(survival_times, censored)
        ax.plot(stimes / 365, sfracs)
        
        ax.set_xlabel('Survival time (year)')
        ax.set_ylabel('Survival rate')
        ax.legend(cluster_names)
```

# Chapter 2 Quantile Normalization

```python
import numpy as np
from scipy import stats

def quantile_norm(X):
    """ normalization of X row for each same distribution
    
    M gene expression matrix of N samples is given
    qunatile normalization have same data range for every sample
    
    calculate average of each column to get average row
    quantile of each row is substituted with each quantile
    
    parameter
    -------------
    X: floating point 2D array (M, N)
        M: gene/properties (column)
        N: sample (row)
    
    return value
    -------------
    Xn: floating point 2D array (M, N)
        normalized data
    """
    
    # Calculation of qunatiles
    quantiles = np.mean(np.sort(X, axis=0), axis=1)
    
    # Rank calculation for each row
    # Calculate each row's rank for each measurement value
    # Smallest value becomes 1, next smallest value becomes 2, ..., Most value becomes M
    ranks = np.apply_along_axis(stats.rankdata, 0, X)
    
    # Change index to integer 1 to M-1
    rank_indices = ranks.astype(int) - 1
    
    # Indexing quantile about each rank of rank matrix
    Xn = quantiles[rank_indices]
    
    return(Xn)
```

```python
def quantile_norm_log(X):
    logX = np.log(X + 1)
    logXn = quantile_norm(logX)
    return logXn
```

## 2.1 Import Data

```python
import numpy as np
import pandas as pd

# Import data TCGA melanoma
filename = 'data/counts.txt'
data_table = pd.read_csv(filename, index_col=0) # parsing pandas

print(data_table.iloc[:5, :5])
```

```python
# 2D ndarray include gene expression of each species
counts = data_table.values
```

```python
# Display graph and apply customized style
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('style/elegant.mplstyle')
```

## 2.2 Difference of gene expression distribution among species

```python
from scipy import stats

def plot_col_density(data):
    # Use kernel density estimation
    density_per_col = [stats.gaussian_kde(col) for col in data.T]
    x = np.linspace(np.min(data), np.max(data), 100)
    
    fig, ax = plt.subplots()
    for density in density_per_col:
        ax.plot(x, density(x))
    ax.set_xlabel('Data value (per column)')
    ax.set_ylabel('Density')
```

```python
# Before normalization
log_counts = np.log(counts + 1)
plot_col_density(log_counts)
```

```python
# After normalization
log_counts_normalized = quantile_norm_log(counts)

plot_col_density(log_counts_normalized)
```
